```python
# import class from crud module
from crudModule import AnimalShelter

# instantiate instance of class
print('Connecting to database...')
shelter = AnimalShelter()
print('Connection Successful')
```

```python
# create a new record in the database
test_animal = {
      'rec_num': 11,
      'age_upon_outcome': '1 year',
      'animal_id': 'TestAnimal',
      'animal_type': 'Dog',
      'breed': 'Dachshund Wirehair Mix',
      'color': 'Tan/White',
      'date_of_birth': '2015-02-23',
      'datetime': '2025-11-22 17:49:00',
      'monthyear': '2025-11-22T17:49:00',
      'name': 'Salamander',
      'outcome_subtype': '',
      'outcome_type': 'Adoption',
      'sex_upon_outcome': 'Spayed Female',
      'location_lat': 30.7290272761146,
      'location_long': -97.3753328216134,
      'age_upon_outcome_in_weeks': 52.8203373015873
    }

print('Inserting test animal')
result = shelter.create(test_animal)
if result:
    print('Create Successful')
else:
    print('Create Failed')
```

```python
print('Testing Read Method - Single Record')
read_query = {'animal_id' : 'TestAnimal'}
result = shelter.read(read_query)
if result:
    print(f"Read successful! Found {len(result)} record(s)")
    for animal in result:
        print(f"Name: {animal['name']}")
        print(f"Breed: {animal['breed']}")
        print(f"Animal ID: {animal['animal_id']}")
else:
    print("Read failed or no results found")
```

```python
print('Testing Read Method - Multiple Records')
read_query2 = {'breed' : 'German Shepherd'}
result = shelter.read(read_query2)
if result:
    print(f"Read successful! Found {len(result)} record(s)")
    for animal in result:
        print(f"Name: {animal['name']}")
        print(f"Breed: {animal['breed']}")
else:
    print("Read failed or no results found")
```
